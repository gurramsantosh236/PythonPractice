```python
from collections import deque

# Example usage
d = deque([1, 2, 3])
print(f'intial d list:{d}')
d.append(4)
print(f"After append(4): {d}")
```

```python
d.appendleft(0)
print(f"After appendleft(0): {d}")
```

```python
right_item = d.pop()
print(f"Popped right: {right_item}, Deque: {d}")
```

```python
left_item = d.popleft()
print(f"Popped left: {left_item}, Deque: {d}")
```

```python
d.rotate(1)
print(f"After rotate(1): {d}")
```
