```python
#sorting the list using loop
list2=[4,5,2,9,1,3,8,7,6,0]
print(list2)
for i in range(len(list2)):
  for j in range(i+1,len(list2)):
    if list2[i]>list2[j]:
      list2[i],list2[j]=list2[j],list2[i]

print(list2)


```

```python
#sorting the list using while loop
list2=[4,5,2,9,1,3,8,7,6,0]
print(list2)
i=0

while i < len(list2):
  j=i+1
  while j<len(list2):
    if list2[i]>list2[j]:
      list2[i],list2[j]=list2[j],list2[i]
    j+=1
  i+=1

print(list2)

```

```python
#Left Trianle pattern

n=5
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end=' ')
  print()
```

```python
# inverted left trangle
n=5
for i in range(n,0,-1):
  for j in range(i,0,-1):
    print(j,end=' ')
  print()
```

```python
#Floyd's Triangle
n=5
num=1
for i in range(1,n+1):
  for j in range(1,i+1):
    print(num,end=" ")
    num+=1
  print()
```

```python
# pyramid number pattern
def pyramid_number(n):
  for i in range(1,n+1):
    print(" "*(n-i),end='')
    for j in range(1,i+1):
      print(j,end=' ')
    print()

pyramid_number(10)
```

```python
#Reverse number triangle
n=5
for i in range(n,0,-1):
  print(" "*(n-1),end="")
  for j in range(n,i-1,-1):
    print(i,end=' ')
  print()
```

```python
import numpy as np

# range for floats with np.arange()
for i in np.arange(0, 4.5, 0.5):
    print(i, end=', ')
# Output 0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0,
print()
# Example 2
for i in np.arange(5.5, 15.5, 2.5):
    print(i, end=' ')
# Output 5.5, 8.0, 10.5, 13.0,
```

```python
import numpy as np

# reverse range of floats
for i in reversed(np.arange(5.5, 30.5, 5.5)):
    print(i, end=', ')
```

```python
import numpy as np

# Float range using np.linspace()
# from 2.5 to 12.5
# num = total float numbers in the output
for i in np.linspace(2.5, 12.5, num=5):
    print(i, end=', ')
# Output 2.5, 5.0, 7.5, 10.0, 12.5,
print('')

# endpoint=False to not include stop number in the result
for i in np.linspace(2.5, 12.5, num=5, endpoint=False):
    print(i, end=', ')
# Output 2.5, 4.5, 6.5, 8.5, 10.5,
```
