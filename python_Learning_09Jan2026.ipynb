```python
#sorting the list using loop
list2=[4,5,2,9,1,3,8,7,6,0]
print(list2)
for i in range(len(list2)):
  for j in range(i+1,len(list2)):
    if list2[i]>list2[j]:
      list2[i],list2[j]=list2[j],list2[i]

print(list2)


```

```python
#sorting the list using while loop
list2=[4,5,2,9,1,3,8,7,6,0]
print(list2)
i=0

while i < len(list2):
  j=i+1
  while j<len(list2):
    if list2[i]>list2[j]:
      list2[i],list2[j]=list2[j],list2[i]
    j+=1
  i+=1

print(list2)

```
