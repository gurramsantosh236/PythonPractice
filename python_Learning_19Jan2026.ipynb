```python
#bubble sorting
lst=[3,60,2,6,9,1,0,22,11,8]

for i in range(len(lst)):
  for j in range(i+1,len(lst)):
    if lst[j]<lst[i]:
      lst[i],lst[j]=lst[j],lst[i]


print(lst)
lst.reverse()
print(lst)

```

```python
fruits=['apple',"banana",'cherry']
veggies=['carrot','beans']
fruits.append("date")
print("append:",fruits)
fruits.extend(veggies)
print("extend:",fruits)
fruits.insert(1,'grapes')
print("insert:",fruits)
fruits.pop(1)
print("pop:",fruits)
fruits[0]="Gova"
print("update:",fruits)
fruits.remove('cherry')
print("remove:",fruits)
fruits.clear()
print("clear:",fruits)
```

```python
import turtle
import math

t=turtle.Turtle()
t.speed(0)
t.color("red")
turtle.bgcolor("black")

def carozon(n):
  x=16*math.sin(n)**3
  y=13*math.cos(n)-5*math.cos(2*n)-2*math.cos(3*n)-math.cos(4*n)
  return x,y
t.penup()
for i in range(15):
  t.goto(0,0)
  t.pendown()
  for n in range(0,100,2):
    x,y=carozon(n/10)
    t.goto(x*i,y*i)
  t.penup()

t.hideturtle()
turtle.done()
```
