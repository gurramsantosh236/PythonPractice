```python
#bubble sorting
lst=[3,60,2,6,9,1,0,22,11,8]

for i in range(len(lst)):
  for j in range(i+1,len(lst)):
    if lst[j]<lst[i]:
      lst[i],lst[j]=lst[j],lst[i]


print(lst)
lst.reverse()
print(lst)

```
