```python
import random
calulate_number=random.randint(1,10)
attempt=0
print(calulate_number)
print("Guess the number between 1 to 10")

while True and attempt<5:
  try:
    print("Attempt:",attempt+1)
    attempt+=1
    userinput=int(input("Enter your Number:"))
    #userinput=int(input("Enter your Number:"))
    if userinput>calulate_number:
      print("Entered number is greater than guessed number,try again")
    elif userinput<calulate_number:
      print("Entered number is smaller than guessed number,try again")
    else:
      print("Congratulations,You have guessed the correct number")
      break
  except ValueError:
    print("Invalid input.Please enter a valid number.")






```
