```python
import random
calulate_number=random.randint(1,10)
attempt=0
print(calulate_number)
print("Guess the number between 1 to 10")

while True and attempt<5:
  try:
    print("Attempt:",attempt+1)
    attempt+=1
    userinput=int(input("Enter your Number:"))
    #userinput=int(input("Enter your Number:"))
    if userinput>calulate_number:
      print("Entered number is greater than guessed number,try again")
    elif userinput<calulate_number:
      print("Entered number is smaller than guessed number,try again")
    else:
      print("Congratulations,You have guessed the correct number")
      break
  except ValueError:
    print("Invalid input.Please enter a valid number.")






```

```python
# Verify the strength of the password

password = input('Please enter your password: ')
is_long=len(password)>=8
has_upper=any(c.isupper() for c in password)
has_lower=any(c.islower() for c in password)
has_digit=any(c.isdigit() for c in password)
has_special=any(not c.isalnum() for c in password)
print(is_long)
print(has_upper)
print(has_lower)
print(has_digit)
print(has_special)

lst=[]
for char in password:
  lst.append(char)
print(lst)

lst2=list(password)
print(lst2)
lst3=[]
lst3.extend(password)
print(lst3)

```
